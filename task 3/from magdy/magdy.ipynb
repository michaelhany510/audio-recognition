```python
import librosa
import librosa.display
import IPython.display as ipd
import os
import numpy as np
import itertools
import pandas as pd
```

```python
def feature_extraction1(file_path):
    # load the audio file
    x, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    # extract features from the audio
    mfcc = np.mean(librosa.feature.mfcc(y=x, sr=sample_rate, n_mfcc=13).T, axis=0)

    return mfcc
```

```python
file_path = './files/magdy/15.wav'
mfcc = feature_extraction1(file_path)
print(mfcc.T)
print(mfcc.shape)
```

```python
def feature_extraction2(file_path):
    # load the audio file
    x, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    # extract features from the audio
    tempo = librosa.beat.tempo(x)[0]

    return tempo
```

```python
def feature_extraction3(file_path):
    # load the audio file
    x, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    # extract features from the audio
    zero_crossings = sum(librosa.zero_crossings(x, pad=False))

    return zero_crossings
```

```python
features1magdy = []
features2magdy = []
features3magdy = []
directory = './files/magdy/'
for audio in os.listdir('./files/magdy/'):
    audio_path = directory + audio
    features1magdy.append(feature_extraction1(audio_path))
    features2magdy.append(feature_extraction2(audio_path))
    features3magdy.append(feature_extraction3(audio_path))
```

```python
features1mayar = []
features2mayar = []
features3mayar = []
directory = './files/mayar/'
for audio in os.listdir('./files/mayar/'):
    audio_path = directory + audio
    features1mayar.append(feature_extraction1(audio_path))
    features2mayar.append(feature_extraction2(audio_path))
    features3mayar.append(feature_extraction3(audio_path))
```

```python
features1mostafa = []
features2mostafa = []
features3mostafa = []
directory = './files/mostafa/'
for audio in os.listdir('./files/mostafa/'):
    audio_path = directory + audio
    features1mostafa.append(feature_extraction1(audio_path))
    features2mostafa.append(feature_extraction2(audio_path))
    features3mostafa.append(feature_extraction3(audio_path))
```

```python
features1magdy   = list(itertools.chain.from_iterable(features1magdy))
features1mayar   = list(itertools.chain.from_iterable(features1mayar))
features1mostafa = list(itertools.chain.from_iterable(features1mostafa))
# features1mayar
# type(features1mayar[0])
```

```python
dfmagdy = pd.DataFrame()
dfmagdy['f1'] = features1magdy
dfmagdy['f2'] = features2magdy
dfmagdy['f3'] = features3magdy
dfmagdy['target'] = [0] * 15
dfmagdy
```

```python
dfmayar = pd.DataFrame()
dfmayar['f1'] = features1mayar[:15]
dfmayar['f2'] = features2mayar[:15]
dfmayar['f3'] = features3mayar[:15]
dfmayar['target'] = [1] * 15
dfmayar
```

```python
dfmostafa = pd.DataFrame()
dfmostafa['f1'] = features1mostafa[:15]
dfmostafa['f2'] = features2mostafa[:15]
dfmostafa['f3'] = features3mostafa[:15]
dfmostafa['target'] = [2] * 15
dfmostafa
```

```python
df = pd.concat([dfmagdy, dfmayar, dfmostafa], ignore_index=True)
df
```

```python
#Import Libraries
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
#----------------------------------------------------

#X Data
x = df[['f1', 'f2', 'f3']]

#y Data
y = df['target']

#----------------------------------------------------
#Splitting data

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=44, shuffle =True)

#----------------------------------------------------
#Applying SVR Model 

'''
sklearn.svm.SVR(kernel=’rbf’, degree=3, gamma=’auto_deprecated’, coef0=0.0, tol=0.001,
                C=1.0, epsilon=0.1, shrinking=True, cache_size=200, verbose=False,max_iter=-1)
'''

SVRModel = SVR(C = 150 , epsilon=0.1 ,kernel = 'rbf') # it also can be : linear, poly, rbf, sigmoid, precomputed
SVRModel.fit(X_train, y_train)

#Calculating Details
print('SVRModel Train Score is : ' , SVRModel.score(X_train, y_train))
print('SVRModel Test Score is  : ' , SVRModel.score(X_test, y_test))
print('----------------------------------------------------')
```
